What crimes occur with the highest frequency?
Let's find these crime categories and plot the top five results.

```python
import pandas as pd
import matplotlib.pyplot as plt

#load data into dataframre
df = pd.read_csv("Crime_Data_from_2010_to_2019_20240421.csv")

#Count top five crime occurences using column "Crm Cd"
top_crimes = (df['Crm Cd Desc'].value_counts()).iloc[:5]

#Plot the top five crime occurences
top_crimes.plot(kind='barh', edgecolor='black')

#Label the plot axis
plt.xlabel('Number of Reports')
plt.ylabel('Crime Code')
```

Let's take the top five crimes reported and find which locations have their highest frequency.
We will used the column 'Area' for location instead of a street address.  This gives us a police station's geographic area so we can adjust the budgeting for each police station if needed.

```python
#Select rows that only include each Crime Code we are interested in
df624 = df.loc[df['Crm Cd Desc'] == 'BATTERY - SIMPLE ASSAULT']
df330 = df.loc[df['Crm Cd Desc'] == 'BURGLARY FROM VEHICLE']
df510 = df.loc[df['Crm Cd Desc'] == 'VEHICLE - STOLEN']
df440 = df.loc[df['Crm Cd Desc'] == 'THEFT PLAIN - PETTY ($950 & UNDER)']
df310 = df.loc[df['Crm Cd Desc'] == 'BURGLARY']
#The numbers are the 3 digit codes used in the UCR Reporting pdf, also listed under "Crm Cd"
```

Let's find the locations with the highest reporting frequency for each crime

```python
#Count top three "Battery Simple Assault" crime locations using column "Area"
top_location_624 = (df624['AREA NAME'].value_counts()).iloc[:3]

#Plot the top three crime occurences
top_location_624.plot(kind='barh', edgecolor='black', rot=0)

#Label the plot axis
plt.xlabel('Number of Reports')
plt.ylabel('Battery Simple Assault - Crime Area')
```

```python
#Count top three "BURGLARY FROM VEHICLE" crime locations using column "Area"
top_location_330 = (df330['AREA NAME'].value_counts()).iloc[:3]

#Plot the top three crime occurences
top_location_330.plot(kind='barh', edgecolor='black', rot=0)

#Label the plot axis
plt.xlabel('Number of Reports')
plt.ylabel('BURGLARY FROM VEHICLE - Crime Area')
```

```python
#Count top three "VEHICLE - STOLEN" crime locations using column "Area"
top_location_510 = (df510['AREA NAME'].value_counts()).iloc[:3]

#Plot the top three crime occurences
top_location_510.plot(kind='barh', edgecolor='black', rot=0)

#Label the plot axis
plt.xlabel('Number of Reports')
plt.ylabel('VEHICLE - STOLEN - Crime Area')
```

```python
#Count top three "THEFT PLAIN - PETTY ($950 & UNDER)" crime locations using column "Area"
top_location_440 = (df440['AREA NAME'].value_counts()).iloc[:3]

#Plot the top three crime occurences
top_location_440.plot(kind='barh', edgecolor='black', rot=0)

#Label the plot axis
plt.xlabel('Number of Reports')
plt.ylabel('THEFT PLAIN - PETTY ($950 & UNDER) - Crime Area')
```

```python
#Count top three "BURGLARY" crime locations using column "Area"
top_location_310 = (df310['AREA NAME'].value_counts()).iloc[:3]

#Plot the top three crime occurences
top_location_310.plot(kind='barh', edgecolor='black', rot=0)

#Label the plot axis
plt.xlabel('Number of Reports')
plt.ylabel('BURGLARY - Crime Area')
```

We now know the highest reported categorical crimes and their highest reported locations.  This gives us a good start to improve any budgetary plans for police presence in those areas.
